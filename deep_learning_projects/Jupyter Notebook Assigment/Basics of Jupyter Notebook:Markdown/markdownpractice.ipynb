## Headers in Markdown


We use the key "#" to get the different headers.
 For example, to get this header:

## Math Symbols

Math symbols in markdown are imported from the [LaTeX language](https://www.latex-project.org/about/).

One can work with math symbols *inline* (in the middle of text) or in its exclusive space (*display math*).

Inline math equation: $\omega = d\phi / dt$. 

Display math should get its own line like so:

$$I = \int \rho R^{2} dV$$

One can get beautiful math expressions such as:
$$
\displaystyle\sum_{k=3}^5 k^2=3^2 + 4^2 + 5^2-\vec{\alpha}+\frac{\beta}{\gamma} =50
$$

Note that you can backslash-escape any punctuation characters
which you wish to be displayed literally, ex.: $\$$.

There are lots of math symbols that can be used in Markdown. Here is a good [cheatsheet](https://www.upyesp.org/posts/makrdown-vscode-math-notation/) on these symbols.
